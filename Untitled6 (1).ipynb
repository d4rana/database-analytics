```r
library(ggplot2)
```

```r
df <- read.csv('/content/Energy_dataset.csv')
```

```r
library(ggplot2)
ggplot(df, aes(x = Wall.Area, y = Cooling.Load)) +
    geom_point() +
    labs(title = "Scatter Plot", x = "Wall.Area", y = "Cooling.Load")
```

```r
library(dplyr)
library(ggplot2)

tranformed_df <- df %>%
  group_by(Wall.Area) %>%
  summarise(mean_cooling_load = mean(Cooling.Load))

ggplot(tranformed_df, aes(x = Wall.Area, y = mean_cooling_load)) +
  geom_bar(stat = "identity") +
  labs(title = "Bar Plot", x = "Wall.Area", y = "Mean Cooling Load")
```

```r
library(tidyr)
library(ggplot2)

long_df <- df %>%
  gather(key = "area_type", value = "area_value", Wall.Area)

ggplot(long_df, aes(x = area_value, y = Cooling.Load, color = area_type, group = area_type)) +
  geom_line() +
  labs(title = "Line Plot", x = "Area", y = "Cooling Load")
```
